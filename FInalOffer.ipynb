```python
import pandas as pd
from pandas import read_excel

file_name = 'mf_revenue_prediction.xlsx' 
mf_revenue_prediction = read_excel(file_name)

file_name = 'cc_revenue_prediction.xlsx' 
cc_revenue_prediction = read_excel(file_name)

file_name = 'cl_revenue_prediction.xlsx' 
cl_revenue_prediction = read_excel(file_name)

data = mf_revenue_prediction.merge(cc_revenue_prediction, how="outer", on="Client")
data = data.merge(cl_revenue_prediction, how="outer", on="Client")
data = data.fillna(0)
max_probability =  data[["probability_x", "probability_y", "probability"]].max(axis=1)
data["max_probability"]=max_probability
final_offer = data.sort_values("max_probability", ascending=False)
final_offer = final_offer[:100]
print(final_offer)
final_offer.to_excel("final_offer.xlsx", index=False)
```

```python
final_offer
```

```python
mf_offer = final_offer[final_offer.probability_x == final_offer.max_probability]
mf_offer = mf_offer[["Client", "probability_x", "mf_revenue_prediction"]]
mf_offer
```

```python
cc_offer = final_offer[final_offer.probability_y == final_offer.max_probability]
cc_offer = cc_offer[["Client", "probability_y", "cc_revenue_prediction"]]
cc_offer
```

```python
cl_offer = final_offer[final_offer.probability == final_offer.max_probability]
cl_offer = cl_offer[["Client", "probability", "cl_revenue_prediction"]]
cl_offer
```

```python
mf_offer.to_excel("mf_offer.xlsx", index=False)
cc_offer.to_excel("cc_offer.xlsx", index=False)
cl_offer.to_excel("cl_offer.xlsx", index=False)
```
